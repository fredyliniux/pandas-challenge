```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
purchase_data.head()
```

```python
players = len(purchase_data['SN'].unique())

players_df = pd.DataFrame({"Total # of Players": [len(purchase_data.SN.unique())]})
players_df
```

```python
#Unique items
unique_items = len(purchase_data['Item ID'].unique())

#Aaverage price
avg_price= purchase_data['Price'].mean()

#Number of purchases
total_purchases= len(purchase_data['Purchase ID'])

#Total revenue
total_revenue= sum(purchase_data['Price'])

#Daatframe
summary_df = pd.DataFrame({
    'Number of Unique Items': unique_items,
    'Average Price': '${:.2f}'.format(avg_price),
    'Total Purchases': total_purchases,
    'Total Revenue': '${:.2f}'.format(total_revenue)
},index=[0])
summary_df
```

```python
#Unique SN 
gen = purchase_data.drop_duplicates('SN')

#Count Gender
count_gen = gen['Gender'].value_counts()

#Percentage
percent_gen = (gen['Gender'].value_counts()/players)*100

#Dataframe
gen_df = pd.DataFrame({
    'Count': count_gen,
    'Percent of Total Players': percent_gen.map('{:.2f}%'.format)
})
gen_df
```

```python
gen_purchase = purchase_data.groupby('Gender')
gender_purchase_count = gen_purchase['Gender'].count()
gender_purchase_count #test works

#average purchase pirce
gender_purchase_mean = gen_purchase['Price'].mean()
gender_purchase_mean #test works

#total revenue
gender_purchase_total = gen_purchase['Price'].sum()
gender_purchase_total #test works
```

```python
gender_purchase_data_df = pd.DataFrame({
    'Purchase Count': gender_purchase_count,
    'Avg Purchase Price': gender_purchase_mean.map('${:,.2f}'.format),
    'Total Revenue': gender_purchase_total.map('${:,.2f}'.format),
    'Avg Total Purchase per Person': (gender_purchase_total/count_gen).map('${:,.2f}'.format),
})
gender_purchase_data_df
```

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 40]

age_range = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
```

```python
#categorize the existing players using the age bins
purchase_data['age_range'] =  pd.cut(purchase_data['Age'], bins, labels=age_range)
```

```python

#groupby
range_age = purchase_data.groupby('age_range').SN.nunique()

#create dataframe to show total count and percentage of total players for each group
age_df = pd.DataFrame({
    'Total Count': range_age,
    'Percentage of Players': (range_age / purchase_data.SN.nunique() * 100).map("{:.2f}".format).map("{}%".format)
})
age_df
```
